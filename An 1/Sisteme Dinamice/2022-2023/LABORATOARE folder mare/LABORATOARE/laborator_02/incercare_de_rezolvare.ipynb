```python
# 1
# a)
x=var('x')
y=function('y')(x)
ec=diff(y,x)==2*x*(1+y**2)
show(ec)

sol=desolve(ec,y)
show(sol)

_C=var('_C')
sol1=solve(sol,y)
show(sol1)

d=sol1[0].rhs()

d1=d.substitute(_C==1)
g=plot(d1,-10,10,ymin=-5,ymax=5,detect_poles=True)
show(g)
for i in [2..10]:
    g=g+plot(d.substitute(_C==i),-5,5,ymin=-10,ymax=10,detect_poles=True)
show(g)
```

```python
# b)
x=var('x')
y=function('y')(x)
ec=diff(y,x)*(x**2-1)+2*x*y**2==0
show(ec)

sol=desolve(ec,y)
show(sol)

_C=var('_C')
sol1=solve(sol,y)
show(sol1)

d=sol1[0].rhs()

d1=d.substitute(_C==1)
g=plot(d1,2,10,detect_poles=True)

for i in [2..10]:
    g=g+plot(d.substitute(_C==i),2,10,detect_poles=True)
show(g)
```
