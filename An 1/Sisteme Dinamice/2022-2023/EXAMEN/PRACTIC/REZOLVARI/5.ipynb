```python
#scriu ecuatia autonoma
s=var('s')
f=function('f')(s)
f(s)=-s^3+2*s^2+3*s
show(f(s))
```

```python
#aflu punctele de echilibru
sol=solve(f(s)==0,s)
show(sol)
```

```python
#stabilitatea
# daca f'(x*) < 0 => local asimptotic stabil
# daca f'(x*) > 0 => instabil
x1 = sol[0].rhs()
x2 = sol[1].rhs()
x3 = sol[2].rhs()
show("x1:",diff(f,s)(x1)) # local asimptotic stabil
show("x2:",diff(f,s)(x2)) # local asimptotic stabil
show("x3:",diff(f,s)(x3)) # instabil
```
