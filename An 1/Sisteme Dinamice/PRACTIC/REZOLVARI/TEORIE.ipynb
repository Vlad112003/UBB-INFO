### Portretul fazic:

```python
reset()
t=var('t')
x=function('x')(t)
y=function('y')(t)
#daca vrem si o curba (cu albastru)
deq_1=diff(x)==x-y
deq_2=diff(y)==x+y
syst=[deq_1,deq_2]
C1,C2=var("C1,C2")
C1=1
C2=1
sol=desolve_system(syst,[x,y],[0,C1,C2])
sol_x=sol[0].rhs()
sol_y=sol[1].rhs()
#daca vrem si o curba (cu albastru)

f1=function('f1')(x,y)
f2=function('f2')(x,y)
f1(x,y)=x-y
f2(x,y)=x+y
#portretul fazic
g1=plot_vector_field([f1(x,y),f2(x,y)], [x,-100,100], [y,-100,100],color='red' ,aspect_ratio = 1 )
g2=parametric_plot((sol_x(t), sol_y(t)), (t, -5, 5),color='blue')
show(g1+g2)
```
