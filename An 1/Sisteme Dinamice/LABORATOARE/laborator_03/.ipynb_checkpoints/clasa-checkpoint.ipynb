```python
#2
# a)
t,C1,C2 = var('t,C1,C2')
x=function('x')(t)
y=function('y')(t)

ec1=diff(x,t)==x+4*y
ec2=diff(y,t)==x+y
syst=[ec1,ec2]
sol=desolve_system(syst, [x,y],[0,C1,C2])
sol_x(t,C1,C2)=sol[0].rhs()
sol_y(t,C1,C2)=sol[1].rhs()

g1=plot(sol_x(t,1,2),t,-4,4,ymin=-10,ymax=10,color='red')
g2=plot(sol_y(t,1,2),t,-4,4,ymin=-10,ymax=10,color='blue')
show(g1+g2)
```

```python
# d)
t,C1,C2,C3 = var('t,C1,C2,C3')
x=function('x')(t)
y=function('y')(t)
z=function('z')(t)

ec1=diff(x,t)==-x+3*y+3*z+27*t^2
ec2=diff(y,t)==2*x-2*y-5*z+3*t
ec3=diff(z,t)==-2*x+3*y+6*z+3
syst=[ec1,ec2,ec3]
sol=desolve_system(syst, [x,y,z],[0,C1,C2,C3])
sol_x(t,C1,C2,C3)=sol[0].rhs()
sol_y(t,C1,C2,C3)=sol[1].rhs()
sol_z(t,C1,C2,C3)=sol[2].rhs()

g1=plot(sol_x(t,50,-30,26),t,-1,5,ymin=0,ymax=100,color='red')
g2=plot(sol_y(t,50,-30,26),t,-1,5,ymin=0,ymax=100,color='blue')
g3=plot(sol_z(t,50,-30,26),t,-1,5,ymin=0,ymax=100,color='green')
show(g1+g2+g3)
```
