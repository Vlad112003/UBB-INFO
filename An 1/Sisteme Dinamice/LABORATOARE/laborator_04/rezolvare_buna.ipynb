```python
x0, k,t = var('x0, k,t')
x = function ('x')(t)
expr = diff(x, t) == -k * x
solx(t, x0, k,) = desolve(expr, [x, t], ics = [0, x0])
show(solx(t, x0, k))
```

```python
plot([solx(t, 100, i/10) for i in [1..10]], t, 0, 20, ymin = 0, ymax = 100)
```

```python
t12=var('t12')
exprc = solx(t12, x0, k) == x0 / 2
solf(t12) = solve(exprc, k)
calc = solf(t12)[0].rhs()
show(calc)
```

```python
kc = calc.substitute(t12 == 5730)
show(numerical_approx(kc))
```

```python
reset()
t,r,r0,x0,K=var('t,r,r0,x0,K')
x=function('x')(t)
expr1=diff(x,t)==r*x
solx(t,x0,r)=desolve(expr1,[x,t], ics=[0,x0])
show(solx(t,x0,r))
```

```python
expr2=diff(x,t)==r0*x*(1-x/K)
solx2=desolve(expr2,[x,t], ics=[0,x0])
soly2=solx2.log_simplify()
soly3(t,x0,r0,K)=solve(soly2, x)[0].rhs()
show(soly3(t,x0,r0,K))
```
